```python

#########Differencing time series to make it stationary#########
# Importing the modules
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# fixing the size
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Original Series
fig, axes = plt.subplots(2, 2, sharex=True)
axes[0, 0].plot(data.Failed_Gadgets); axes[0, 0].set_title('Original Series')
plot_acf(data.Failed_Gadgets, ax=axes[0, 1])
# 1st Differencing to make stationary time series data
axes[1, 0].plot(data.Failed_Gadgets.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(data.Failed_Gadgets.diff().dropna(), ax=axes[1, 1])
plt.show()
plt.show()
```

```python
####### Finding the order of the AR model ######
#importing modules
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# PACF plot of 1st differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})
# fixing the axis
fig, axes = plt.subplots(1, 2, sharex=True)
# plotting on differen axis
axes[0].plot(data.Failed_Gadgets.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,5))
# plotting partial autocorrelation function
plot_pacf(data.Failed_Gadgets.diff().dropna(), ax=axes[1])
plt.show()
```

```python
######## Finding the order of the MA model #######
# setting the size
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})
#  fixing the subplots
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(data.Failed_Gadgets.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,1.2))
# plotting the autocorrelation function
plot_acf(data.Failed_Gadgets.diff().dropna(), ax=axes[1])
plt.show()
```

```python
# importing the ARIMA model
from statsmodels.tsa.arima_model import ARIMA
# 1,1,1 ( arima p d q )
model = ARIMA(data.Failed_Gadgets, order=(0,0,0))
# Training arima modeling
model_fit = model.fit()
```

```python
# arima model results
model_fit.plot_predict(dynamic=False)
plt.show()
```
